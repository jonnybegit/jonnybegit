```python
import pandas as pd
```

```python
mpp = pd.read_excel("C:\Work\Datasets\DigestofUKEnergyStatistics\DUKES_5.11_Powerplants.xls", sheet_name='5.11', header=4, nrows=409)
mpp['Installed Capacity\n(MW)'] = pd.to_numeric(mpp['Installed Capacity\n(MW)'], downcast="float")
mpp.info()
```

```python
mpp.head()
```

```python
company_stats = mpp.groupby('Company Name').sum()
company_stats.sort_values('Installed Capacity\n(MW)', ascending=False)

```

```python
technology_stats = mpp.groupby(['Fuel', 'Technology']).sum()
technology_stats.sort_values('Installed Capacity\n(MW)', ascending=False)
```
